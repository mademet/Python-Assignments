```python
a = int(input("Please enter a number:"))
c = True
if a <= 1:
    c = False
else:
    for i in range (a-2):
        if a%(i+2) == 0:
            c = False
            break
if c:
    print(a, "is a prime number.")
else:
    print(a, "is not a prime number.")
```

```python
a = int(input("Please enter a number bigger than 1: "))
l = []
for i in range (2, (a + 1)):
    c = True
    for ii in range (2, i):
        if i%(ii) == 0:
            c = False
            break
    if c:
        l.append(i)
print("the prime numbers 1 to", a,": ",l)
```
