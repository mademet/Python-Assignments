```python
chars = chr(random.randint(33, 47)) + chr(random.randint(58, 64))
```

```python
password = "".join(uppers) +  "".join(lowers) +  "".join(numbers) +  "".join(chars)
```

```python
def shuffleit(string):
    l = list(string)
    random.shuffle(l)
    return "".join(l)
```

```python
def for_loop():
    result = []    
    for i in range(1000000):
        result.append(i)
    return result
```

```python
def for_comp():
    return [i for i in range(1000000)]
```

```python
from timeit import timeit
```

```python
time_1 = timeit(for_loop, number = 100 )
time_2 =timeit(for_comp, number = 100 )
```

```python
print(time_1 / time_2)
```
